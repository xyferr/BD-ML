#DATA VISUALISATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(top_10_cf['Country/Region'],top_10_cf['Confirmed'],label = 'Confirmed')
```

```python
plt.figure(figsize=(10,6))
plt.plot(top_10_cf['Country/Region'],top_10_cf['Deaths'],label = 'Deaths')
```

```python

plt.figure(figsize=(10, 10))

# Plot 'Confirmed' and 'Deaths' on the same graph
plt.plot(top_10_cf['Country/Region'], top_10_cf['Confirmed'], label='Confirmed', marker='o', linestyle='-', color='blue')
plt.plot(top_10_cf['Country/Region'], top_10_cf['Deaths'], label='Deaths', marker='x', linestyle='-', color='red')

# Adding labels and title
plt.xlabel('Country/Region')
plt.ylabel('Count')


# Adding a legend
plt.legend()



# Show the plot
plt.tight_layout() # Optional: Adjust layout for better display
plt.show()
```
