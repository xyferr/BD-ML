```python


# Extract data for 'Confirmed' and 'Deaths' from your DataFrame
Confirmed = top_10_cf['Confirmed']
Deaths = top_10_cf['Deaths']
country = top_10_cf['Country/Region']

# Calculate the x positions for the bars
xpos = np.arange(len(country))

# Create the bar chart
plt.bar(xpos-0.2, Confirmed, width=0.4, label="Confirmed Cases")
plt.bar(xpos+0.2, Deaths, width=0.4, label="Deaths")
plt.yscale('log')  # This will change the y-axis scale to a logarithmic scale, which is particularly useful when there's a large disparity between values.

# Set the x-tick labels to be the country names
plt.xticks(xpos, country, rotation=90)  # Rotating labels for readability

plt.ylabel("Count")
plt.title('COVID-19 Confirmed Cases and Deaths by Country/Region')
plt.legend()
plt.tight_layout()  # Ensuring the labels fit within the plot

plt.show()
```

PIE CHARTS

```python

# Define the number of top countries you want to consider
top_n_countries = 10

# Sort the DataFrame by 'Confirmed' cases in descending order
cov_sorted = cov.sort_values(by='Confirmed', ascending=False)

# Get the top 20 countries with the highest confirmed cases
top_countries_confirmed = cov_sorted.head(top_n_countries)

# Create a pie chart for the top 20 countries with confirmed cases
data_confirmed = top_countries_confirmed['Confirmed'].tolist()
labels_confirmed = top_countries_confirmed['Country/Region'].tolist()
# Aggregate the remaining cases into the 'Other' category
other_confirmed = cov_sorted.iloc[top_n_countries:]['Confirmed'].sum()
data_confirmed.append(other_confirmed)
labels_confirmed.append('Other')

# Create a pie chart for confirmed cases
plt.figure(figsize=(8, 8))
plt.pie(data_confirmed, labels=labels_confirmed, autopct='%1.1f%%', startangle=140)
plt.title('Confirmed Cases Distribution (Top 10 Countries)')
plt.axis('equal')
plt.show()

# Sort the DataFrame by 'Deaths' in descending order
cov_sorted = cov.sort_values(by='Deaths', ascending=False)

# Get the top 20 countries with the highest deaths
top_countries_deaths = cov_sorted.head(top_n_countries)

# Create a pie chart for the top 20 countries with deaths
data_deaths = top_countries_deaths['Deaths'].tolist()
labels_deaths = top_countries_deaths['Country/Region'].tolist()
# Aggregate the remaining deaths into the 'Other' category
other_deaths = cov_sorted.iloc[top_n_countries:]['Deaths'].sum()
data_deaths.append(other_deaths)
labels_deaths.append('Other')

# Create a pie chart for deaths
plt.figure(figsize=(8, 8))
plt.pie(data_deaths, labels=labels_deaths, autopct='%1.1f%%', startangle=140)
plt.title('Deaths Distribution (Top 10 Countries)')
plt.axis('equal')
plt.show()

# Sort the DataFrame by 'Recovered' cases in descending order
cov_sorted = cov.sort_values(by='Recovered', ascending=False)

# Get the top 20 countries with the highest recoveries
top_countries_recovered = cov_sorted.head(top_n_countries)

# Create a pie chart for the top 20 countries with recoveries
data_recovered = top_countries_recovered['Recovered'].tolist()
labels_recovered = top_countries_recovered['Country/Region'].tolist()
# Aggregate the remaining recoveries into the 'Other' category
other_recovered = cov_sorted.iloc[top_n_countries:]['Recovered'].sum()
data_recovered.append(other_recovered)
labels_recovered.append('Other')

# Create a pie chart for recoveries
plt.figure(figsize=(8, 8))
plt.pie(data_recovered, labels=labels_recovered, autopct='%1.1f%%', startangle=140)
plt.title('Recovered Cases Distribution (Top 10 Countries)')
plt.axis('equal')
plt.show()
```

```python
# Extract data for confirmed cases
confirmed_data = cov.groupby('Country/Region')['Confirmed'].sum()

# Select the top N countries with the highest confirmed cases
top_n = 10  # Change this to your desired number
top_countries = confirmed_data.sort_values(ascending=False).head(top_n)

# Create a subplot grid
fig, axes = plt.subplots(2, 5, figsize=(15, 7))  # Change the number of rows and columns as needed

# Iterate through the top countries and create pie charts
for i, (country, cases) in enumerate(top_countries.items()):
    ax = axes[i // 5, i % 5]
    ax.pie([cases, confirmed_data.sum() - cases], labels=[country, 'Other'], autopct='%1.1f%%', startangle=140)
    ax.set_title(f'{country}')

# Use tight_layout to improve spacing
plt.tight_layout()
plt.show()
```

```python
numeric_columns = cov.select_dtypes(include=[np.number])  # Select only numeric columns
corr_matrix = numeric_columns.corr()  # Calculate the correlation matrix

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Set the threshold for correlations to be displayed
threshold = 0.5  # You can adjust this threshold as needed

# Create a mask for correlations above the threshold
mask = np.abs(corr_matrix) >= threshold

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap with the masked correlation matrix
sns.heatmap(corr_matrix[mask], annot=True, cmap='coolwarm', linewidths=0.5)

# Set the title for the heatmap
plt.title(f'Correlation Heatmap (Threshold: {threshold})')

# Display the heatmap
plt.show()
```
